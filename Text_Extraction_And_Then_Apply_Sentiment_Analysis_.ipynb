```python
import pandas as pd

# Load the Excel file
df = pd.read_excel("/content/pdf_text_with_coordinates.xlsx")

# Strip any leading/trailing spaces in column names
df.columns = df.columns.str.strip()

# Show the column names
print("Columns:", df.columns.tolist())

# Preview the data
print(df.head())
```

```python
from textblob import TextBlob

# Ensure 'text' column is string type
df['Text'] = df['Text'].astype(str)

# Define a function to get polarity score
def get_polarity(Text):
    return TextBlob(Text).sentiment.polarity

# Apply the function to compute polarity
df['Polarity'] = df['Text'].apply(get_polarity)

# Add sentiment label based on polarity
def label_sentiment(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df['Sentiment'] = df['Polarity'].apply(label_sentiment)
```

```python
# Save to a new file
output_path = "/content/pdf_text_with_coordinates.xlsx"
df.to_excel(output_path, index=False)

print(f"✅ Sentiment analysis completed. File saved to: {output_path}")
```
